```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new['tags']).toarray()
```

```python
vectors.shape
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
#to remove punctuation marks and symbol from name column
import string
new['name'] = new['name'].str.replace(r'[^\w\s]+', '')
    
```

```python
new.head()
```

```python
new[new['name']=='Bloodivores'].index[0]
```

```python
def recommend(movie):
    an_index = new[new['name'] == movie].index[0]
    distances = similarity[an_index]
    anime_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x: x[1])[1:6]
    
    for i in anime_list:
        print(new.name.iloc[i[0]])
```

```python
recommend('One Piece')
```
